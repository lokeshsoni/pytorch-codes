```python
from fastai import accuracy

from fastai.vision import (
    ImageDataBunch, 
    get_transforms, 
    imagenet_stats, 
    create_cnn,
    models,
    open_image
)
```

```python
path = "tmp/"
```

```python
data = ImageDataBunch.from_folder(path, ds_tfms=get_transforms(), size=224).normalize(imagenet_stats)
```

```python
learn = create_cnn(data, models.resnet34, metrics=accuracy)
```

```python
learn.load("/home/jupyter/.fastai/data/dogscats/models/dogs_cats_resnet34")
```

```python
def predict(img_path):
    img = open_image(img_path)
    prediction = learn.predict(img)
    
    return {"cat" : round(prediction[2][0].item() * 100, 2), "dog" : round(prediction[2][1].item() * 100, 2)}
```

```python
img_path = "/home/jupyter/.fastai/data/dogscats/valid/cats/cat.1001.jpg"
predict(img_path)
```

```python
img_path = "/home/jupyter/.fastai/data/dogscats/valid/dogs/dog.10001.jpg"
predict(img_path)
```

```python
img_path = "/home/jupyter/.fastai/data/dogscats/test1/10198.jpg"
predict(img_path)
```
